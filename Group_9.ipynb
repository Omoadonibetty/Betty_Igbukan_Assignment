# **top 5 customers based on total sales**

```python
# Calculating and displaying the top 5 customers based on total sales in the DataFrame "autosales"
# Grouping the DataFrame by customer name, summing the sales for each customer, and sorting in descending order.
customer_sales = autosales.groupby("CUSTOMERNAME")["SALES"].sum()
top_5_customers = customer_sales.sort_values(ascending=False).head(5)

# Printing the top 5 customers by sales
print(top_5_customers)
```

# **top 5 cities based on total sales**

```python
# Calculating and displaying the top 5 cities based on total sales in the DataFrame "autosales"
# Grouping the DataFrame by city, summing the sales for each city, and sorting in descending order.
sales_city = autosales.groupby("CITY")["SALES"].sum()
top_5_city = sales_city.sort_values(ascending=False).head(5)

# Printing the top 5 cities by sales
print(top_5_city)
```

```python
# Creating a bar plot to visualize total sales for the top 5 cities in the DataFrame "autosales"
# Using top_5_city Series obtained from grouping the DataFrame by city and calculating the sum of sales.
top_5_city.plot.bar()

# Adding a legend to the plot
plt.legend(loc='upper right')

# Adding a title to the plot
plt.title("Sales across Top 5 Cities", fontdict={"fontsize": 20})
```

Madris is one of the top 5 cities with the highest sales. This might be due to the high demand for automobiles in the city. Paris is number 5 in sales.

# **sales of the top 5 customers across different countries; Most Loyal Customer**

```python
# Calculating and displaying the top 5 customers in each country based on total sales in the DataFrame "autosales"
# Grouping the DataFrame by country and customer name, summing the sales for each combination.
# Getting the top 5 customers in each country based on total sales.
top_customers_by_country = autosales.groupby(['COUNTRY', 'CUSTOMERNAME'])['SALES'].sum().reset_index()

# Using groupby and nlargest to get the top 5 customers in each country based on total sales
top_5_customers_by_country = top_customers_by_country.groupby('COUNTRY').apply(lambda x: x.nlargest(5, 'SALES')).reset_index(drop=True)

# Sorting the overall result to obtain the top 5 customers across all countries
top_5 = top_5_customers_by_country.sort_values(by="SALES", ascending=False).head(5)

# Printing the top 5 customers by country
print(top_5)
```

```python

# Creating a bar plot to visualize the sales of the top 5 customers across different countries in the DataFrame "top_5"
# Using Seaborn's sns.barplot with Sales on the x-axis, Customer Name on the y-axis, and Country as hue.
sns.barplot(data=top_5, x='SALES', y='CUSTOMERNAME', hue='COUNTRY', palette='plasma')

# Adding a legend to the plot
plt.legend(loc='lower right')

# Adding a title to the plot
plt.title("Most Loyal Customers", fontdict={"fontsize": 20})

# Show the plot
plt.show()
```

**Euroshopping brought in the most sales of about 700,000 which indicated that they are the most loyal customers while La Rochelle Gifts is the 5th most loyal customer with about 190,000 in sales with their office in France.**

# **total sales for each month**

```python
# Calculating the total sales for each month in the DataFrame "autosales"
# Grouping the DataFrame by month and summing the sales for each month.
monthly_sales = autosales.groupby('MONTH')['SALES'].sum().reset_index()

# Plotting a line chart to visualize the monthly sales trends
plt.figure(figsize=(10, 6))

# Using plt.plot to create a line chart with SALES on the y-axis and MONTH on the x-axis
plt.plot(monthly_sales['SALES'], monthly_sales['MONTH'], marker='o', linestyle='-', color='y')

# Adding a title to the plot
plt.title('Monthly Sales')

# Adding labels to the x and y axes
plt.xlabel('SALES')

# Displaying the line chart
plt.show()
```

# **Monthly Price Trends**

```python
# Calculating the monthly price trends in the DataFrame "autosales"
# Grouping the DataFrame by month and summing the unit prices for each month.
monthly_price = autosales.groupby('MONTH')['PRICEEACH'].sum().reset_index()

# Plotting a line chart to visualize the monthly price trends
plt.figure(figsize=(10, 6))

# Using plt.plot to create a line chart with PRICEEACH on the y-axis and MONTH on the x-axis
plt.plot(monthly_price['PRICEEACH'], monthly_price['MONTH'], marker='o', linestyle='-', color='r')

# Adding a title to the plot
plt.title('Monthly Price Trends')

# Adding labels to the x and y axes
plt.xlabel('PRICEEACH')

# Displaying the line chart
plt.show()
```

# **total sales for each year**

```python
# Calculating the total sales for each year in the DataFrame "autosales"
# Grouping the DataFrame by year and summing the sales for each year.
yearly_sales = autosales.groupby(' YEAR')['SALES'].sum().reset_index()

# Plotting a line chart to visualize the total yearly sales trends
plt.figure(figsize=(10, 6))

# Using plt.plot to create a line chart with YEAR on the x-axis and SALES on the y-axis
plt.plot(yearly_sales[' YEAR'], yearly_sales['SALES'], marker='o', linestyle='-', color='g')

# Adding a title to the plot
plt.title('Total Yearly Sales', fontdict={"fontsize": 20})

# Adding labels to the x and y axes
plt.xlabel('YEAR')

# Displaying the line chart
plt.show()
```

```python
# Calculating the total quantity ordered for each year in the DataFrame "autosales"
# Grouping the DataFrame by year and summing the quantity ordered for each year.
yearly_qty = autosales.groupby(' YEAR')['QUANTITYORDERED'].sum().reset_index()

# Plotting a line chart to visualize the total yearly quantity ordered trends
plt.figure(figsize=(10, 6))

# Using plt.plot to create a line chart with YEAR on the x-axis and QUANTITYORDERED on the y-axis
plt.plot(yearly_qty[' YEAR'], yearly_qty['QUANTITYORDERED'], marker='o', linestyle='-', color='b')

# Adding a title to the plot
plt.title('Total Yearly Quantity Ordered', fontdict={"fontsize": 20})

# Adding labels to the x and y axes
plt.xlabel('YEAR')

# Displaying the line chart
plt.show()
```
